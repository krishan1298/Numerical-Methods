```python
import numpy as np
import matplotlib.pyplot as plt
def runge_kutta(f,f1, a, b, alp1, alp2, h):
    n = int((b-a)/h)
    t=np.arange(a,b+h,h)
    y = np.zeros([n+1,2])
    y[0,0] = alp1
    y[0,1]=alp2
    w=np.zeros([n+1,2])
    k=1
    for i in range(n):
        k1 = h * y[i,1]
        l1 = h * f(t[i], y[i,0], y[i,1])
        k2 = h * (y[i,1]+ 0.5*l1)
        l2 = h * f(t[i] + 0.5*h, y[i,0] + 0.5*k1, y[i,1] + 0.5*l1)
        k3 = h * (y[i,1] + 0.5*l2)
        l3 = h * f(t[i] + 0.5*h, y[i,0] + 0.5*k2, y[i,1] + 0.5*l2)
        k4 = h * (y[i,1] + l3)
        l4 = h * f(t[i] + h, y[i,0] + k3, y[i,1] + l3)
        
        y[i+1,1] = y[i,1] + (1/6) * (l1 + 2*l2 + 2*l3 + l4)
        y[i+1,0] = y[i,0] + (1/6) * (k1 + 2*k2 + 2*k3 + k4)
        
    for i in range(len(y)):
        if np.isclose(t[i],2*np.pi*k):
            w[i+1,1]=y[i,1]
            w[i+1,0]=y[i+1,0]
    return w

f1= lambda t,u1,u2: u2
f5= lambda t,u1,u2:-0.22*u2-np.sin(u1)+2.7*np.cos(t)
a=-30
b=30
alp1=np.pi/5
alp2=0
h=0.1
print(runge_kutta(f5,f1,a,b,alp1,alp2,h))
```
